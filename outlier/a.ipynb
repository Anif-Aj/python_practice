```python
import random



class Player:

    def __init__(self, name):

        self.name = name

        self.points = 0



    def make_guess(self):

        team_a_score = random.randint(0, 4)

        team_b_score = random.randint(0, 4)

        return f"TeamA {team_a_score}- {team_b_score}TeamB"



def calculate_points(guess, real_score):

    guess_scores = [int(score) for score in guess.replace("TeamA ", "").replace("- ", "").replace("TeamB", "").split("-")]

    real_scores = [int(score) for score in real_score.replace("TeamA ", "").replace("- ", "").replace("TeamB", "").split("-")]

    

    if guess_scores == real_scores:

        return 3

    elif guess_scores == real_scores[::-1]:

        return 2

    elif sum(guess_scores) == sum(real_scores):

        return 1

    else:

        return 0



def simulate_turn(players):

    real_score = f"TeamA {random.randint(0, 4)}- {random.randint(0, 4)}TeamB"

    for player in players:

        guess = player.make_guess()

        points = calculate_points(guess, real_score)

        player.points += points

        print(f"{player.name}: {guess} (Real score: {real_score}) - Points: {points}")



def print_leaderboard(players):

    sorted_players = sorted(players, key=lambda x: (-x.points, int(x.name.replace("player", ""))))

    for i, player in enumerate(sorted_players):

        print(f"{i+1}. {player.name}: {player.points} points")



def main():

    players = [Player(f"player{i+1}") for i in range(10)]

    

    for i in range(10):

        print(f"\nTurn {i+1}:")

        simulate_turn(players)

        if (i+1) % 3 == 0:

            print("\nLeaderboard:")

            print_leaderboard(players)

    

    print("\nFinal Leaderboard:")

    print_leaderboard(players)

    champion = max(players, key=lambda x: (x.points, -int(x.name.replace("player", ""))))

    print(f"\nChampion: {champion.name} with {champion.points} points")



if __name__ == "__main__":

    main()


import random



class Player:

    def __init__(self, name):

        # Initialize the player with a name and start with zero points

        self.name = name

        self.points = 0



    def make_guess(self):

        # Generate a random score prediction for both teams (between 0 and 4 goals)

        team_a_score = random.randint(0, 4)

        team_b_score = random.randint(0, 4)

        return f"TeamA {team_a_score}-{team_b_score} TeamB"





def calculate_points(guess, real_score):

    # Extract numeric scores from the guess and real score strings

    guess_scores = [

        int(score)

        for score in guess.replace("TeamA ", "").replace(" TeamB", "").split("-")

    ]

    real_scores = [

        int(score)

        for score in real_score.replace("TeamA ", "").replace(" TeamB", "").split("-")

    ]



    # Award points based on how accurate the player's prediction was

    if guess_scores == real_scores:  # Exact match (correct score for both teams)

        return 3

    elif guess_scores == real_scores[::-1]:  # Reversed team scores

        return 2

    elif sum(guess_scores) == sum(real_scores):  # Correct total goals but wrong team scores

        return 1

    else:

        return 0  # No points if the guess is incorrect





def simulate_turn(players):

    # Simulate a turn by generating the actual match score and evaluating each player's guess

    real_score = f"TeamA {random.randint(0, 4)}-{random.randint(0, 4)} TeamB"

    print(f"\nReal Score: {real_score}")

    

    # Loop through each player, process their guess, and calculate the points they earn

    for player in players:

        guess = player.make_guess()

        points = calculate_points(guess, real_score)

        player.points += points  # Add the points to the player's total

        print(f"{player.name}: {guess} - Points Earned: {points}")





def print_leaderboard(players):

    # Display the leaderboard by sorting players based on their points (descending) 

    # and resolving ties by player number (ascending)

    sorted_players = sorted(

        players, key=lambda x: (-x.points, int(x.name.replace("player", "")))

    )

    

    # Print each player's ranking, name, and total points

    for i, player in enumerate(sorted_players, 1):

        print(f"{i}. {player.name}: {player.points} points")





def main():

    # Create a list of 10 players, naming them player1 to player10

    players = [Player(f"player{i+1}") for i in range(10)]



    # Run 10 turns, where each turn simulates a match and players make guesses

    for i in range(10):

        print(f"\nTurn {i+1}:")

        simulate_turn(players)



        # Display the leaderboard after every 3 turns

        if (i + 1) % 3 == 0:

            print("\nLeaderboard:")

            print_leaderboard(players)



    # Display the final results after all turns are completed

    print("\nFinal Leaderboard:")

    print_leaderboard(players)

    

    # Determine and display the champion, resolving ties by player number

    champion = max(

        players, key=lambda x: (x.points, -int(x.name.replace("player", "")))

    )

    print(f"\nChampion: {champion.name} with {champion.points} points")





if __name__ == "__main__":

    # Run the main function if the script is executed directly

    main()

```

```python
import pygame
import random
# Initialize Pygame

pygame.init()



# Define some colors

WHITE = (255, 255, 255)

BLACK = (0, 0, 0)

YELLOW = (255, 255, 0)



# Set the width and height of each tile

TILE_SIZE = 100



# Set the margin between each tile

MARGIN = 5



# Set the size of the grid

GRID_SIZE = 3



# Set the size of the window

WINDOW_SIZE = GRID_SIZE * (TILE_SIZE + MARGIN) + MARGIN



# Create the window

WINDOW = pygame.display.set_mode((WINDOW_SIZE, WINDOW_SIZE))



# Set the font for the tiles

FONT = pygame.font.SysFont("Arial", 36)



# Create a 2D list to represent the grid

grid = [[i + j * GRID_SIZE + 1 for i in range(GRID_SIZE)] for j in range(GRID_SIZE)]



# Set the last tile to 0 to represent the empty tile

grid[GRID_SIZE - 1][GRID_SIZE - 1] = 0



# Create a variable to store the selected tile

selected_tile = None



# Create a variable to store the game state

game_started = False



# Game loop

while True:

  for event in pygame.event.get():

    if event.type == pygame.QUIT:

      pygame.quit()

      quit()

    elif event.type == pygame.MOUSEBUTTONDOWN:

      # Get the position of the click

      pos = pygame.mouse.get_pos()

      # Convert the position to grid coordinates

      col = pos[0] // (TILE_SIZE + MARGIN)

      row = pos[1] // (TILE_SIZE + MARGIN)

      # Check if the game has started

      if not game_started:

        # Check if the shuffle button was clicked

        if row == GRID_SIZE - 1 and col == GRID_SIZE - 1:

          # Shuffle the grid

          flat_list = [item for sublist in grid for item in sublist]

          random.shuffle(flat_list)

          grid = [flat_list[i:i + GRID_SIZE] for i in range(0, len(flat_list), GRID_SIZE)]

          game_started = True

      else:

        # Check if a tile was clicked

        if grid[row][col] != 0:

          # Check if a tile is already selected

          if selected_tile is None:

            # Select the tile

            selected_tile = (row, col)

          else:

            # Get the position of the selected tile

            sel_row, sel_col = selected_tile

            # Check if the clicked tile is adjacent to the selected tile

            if abs(sel_row - row) + abs(sel_col - col) == 1:

              # Swap the tiles

              grid[sel_row][sel_col], grid[row][col] = grid[row][col], grid[sel_row][sel_col]

              # Check if the game is won

              flat_list = [item for sublist in grid for item in sublist]

              if flat_list == list(range(1, GRID_SIZE * GRID_SIZE)) + [0]:

                print("Congratulations, you won!")

                pygame.quit()

                quit()

              # Deselect the tile

              selected_tile = None

        # Check if the empty tile was clicked

        elif selected_tile is not None:

          # Get the position of the selected tile

          sel_row, sel_col = selected_tile

          # Swap the tiles

          grid[sel_row][sel_col], grid[row][col] = grid[row][col], grid[sel_row][sel_col]

          # Check if the game is won

          flat_list = [item for sublist in grid for item in sublist]

          if flat_list == list(range(1, GRID_SIZE * GRID_SIZE)) + [0]:

            print("Congratulations, you won!")

            pygame.quit()

            quit()

          # Deselect the tile

          selected_tile = None



  # Draw the grid

  WINDOW.fill(BLACK)

  for row in range(GRID_SIZE):

    for col in range(GRID_SIZE):

      # Draw the tile

      pygame.draw.rect(WINDOW, WHITE, (col * (TILE_SIZE + MARGIN) + MARGIN, row * (TILE_SIZE + MARGIN) + MARGIN, TILE_SIZE, TILE_SIZE))

      # Draw the number on the tile

      if grid[row][col] != 0:

        text = FONT.render(str(grid[row][col]), True, BLACK)

        text_rect = text.get_rect(center=(col * (TILE_SIZE + MARGIN) + MARGIN + TILE_SIZE // 2, row * (TILE_SIZE + MARGIN) + MARGIN + TILE_SIZE // 2))

        WINDOW.blit(text, text_rect)

      # Highlight the selected tile

      if (row, col) == selected_tile:

        pygame.draw.rect(WINDOW, YELLOW, (col * (TILE_SIZE + MARGIN) + MARGIN, row * (TILE_SIZE + MARGIN) + MARGIN, TILE_SIZE, TILE_SIZE), 5)



  # Draw the shuffle button

  if not game_started:

    pygame.draw.rect(WINDOW, WHITE, ((GRID_SIZE - 1) * (TILE_SIZE + MARGIN) + MARGIN, (GRID_SIZE - 1) * (TILE_SIZE + MARGIN) + MARGIN, TILE_SIZE, TILE_SIZE))

    text = FONT.render("Shuffle", True, BLACK)

    text_rect = text.get_rect(center=((GRID_SIZE - 1) * (TILE_SIZE + MARGIN) + MARGIN + TILE_SIZE // 2, (GRID_SIZE - 1) * (TILE_SIZE + MARGIN) + MARGIN + TILE_SIZE // 2))

    WINDOW.blit(text, text_rect)



  # Update the window

  pygame.display.update()






```
